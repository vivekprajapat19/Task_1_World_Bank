```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/API_SP.POP.TOTL_DS2_en_csv_v2_131993/API_SP.POP.TOTL_DS2_en_csv_v2_131993.csv', skiprows=4)
```

```python
df.head()
```

```python
print(df.columns)
```

```python
# Select relevant columns: Country Name and population for 2022
df_2022 = df[['Country Name', '2022']].dropna()
```

```python
# Sort by population descending
df_2022 = df_2022.sort_values(by='2022', ascending=False)
```

```python
# Select top 10 most populous countries
top10 = df_2022.head(10)
```

```python
# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='2022', y='Country Name', data=top10, palette='crest')

# Add labels and title
plt.title('Top 10 Most Populous Countries (2022)', fontsize=16)
plt.xlabel('Population')
plt.ylabel('Country')
plt.ticklabel_format(style='plain', axis='x')
plt.tight_layout()
plt.show()
```

```python
# Filter data for a specific country, e.g., India
country_data = df[df['Country Name'] == 'India']
```

```python
# Select data for years 2000 to 2022
years = [str(year) for year in range(2000, 2023)]
population = country_data[years].values.flatten()
```

```python
# Convert population to billions for readability
population = population / 1e9
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.bar(years, population, color='skyblue')
plt.title('Total Population of India (2000â€“2022)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population (in Billions)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
